## 1) Planteamiento del problema

### **Pregunta**
Considere la siguiente función definida en tiempo continuo:

$$
R(\tau)=\sum_{k\in\{-1,0,1\}} A_k \exp\!\left(-\frac{(\tau-\mu_k)^2}{2\sigma_k^2}\right),
$$

donde:

$$
\mu_k = k,
$$

$$
\sigma_k^2=
\begin{cases}
\sigma^2, & k=0,\\[6pt]
\dfrac{\sigma^2}{|k|!}, & k\neq 0,
\end{cases}
$$

y se cumple:

$$
A_k>0,\qquad A_{-1}=A_1.
$$

Se pide analizar si $R(\tau)$ puede ser considerada una **función de autocorrelación válida** de un proceso estocástico **estacionario en sentido amplio (WSS)**.

---

### **Desarrollo (criterios de validez para una autocorrelación WSS)**

Para un proceso WSS real $x(t)$, su autocorrelación

$$
R_x(\tau)=\mathbb{E}\{x(t)\,x(t+\tau)\}
$$

debe cumplir (condiciones necesarias y suficientes más usadas):

1. **Depender solo de $\tau$ (WSS):**  
   Aquí $R(\tau)$ ya está dada solo en función del retardo $\tau$, así que es consistente con WSS.

2. **Simetría hermítica (para proceso real: paridad):**
   - Para procesos reales:  
     $$
     R_x(\tau)=R_x(-\tau)\quad\text{(función par)}.
     $$
   - Para procesos complejos:  
     $$
     R_x(\tau)=R_x^*(-\tau).
     $$

3. **Ser semidefinida positiva (condición clave):**  
   Para cualquier conjunto de tiempos $\{t_i\}$, la matriz de autocorrelación (tipo Toeplitz)
   $$
   \mathbf{R}_{ij} = R_x(t_i-t_j)
   $$
   debe ser **semidefinida positiva**.

   Una forma equivalente (muy práctica) de verificar esto es usando el **Teorema de Wiener–Khinchin / Teorema de Bochner**:

   $$
   S_x(\omega)=\mathcal{F}\{R_x(\tau)\}\ge 0\quad \text{para todo }\omega,
   $$
   donde $S_x(\omega)$ es la **densidad espectral de potencia (PSD)**.

---

### **Reescritura explícita de $R(\tau)$ con $k\in\{-1,0,1\}$**

Como $|1|!=1$, para $k=\pm 1$ se tiene:

$$
\sigma_{1}^2=\sigma_{-1}^2=\frac{\sigma^2}{1!}=\sigma^2,
$$

y para $k=0$:

$$
\sigma_0^2=\sigma^2.
$$

Entonces **todas las gaussianas tienen la misma varianza** $\sigma^2$ en este caso, y:

$$
R(\tau)=A_0 e^{-\frac{\tau^2}{2\sigma^2}}
\;+\;
A_1 e^{-\frac{(\tau-1)^2}{2\sigma^2}}
\;+\;
A_1 e^{-\frac{(\tau+1)^2}{2\sigma^2}}.
$$

---

### **(i) Verificación de paridad $R(\tau)=R(-\tau)$**

- El término central $A_0 e^{-\tau^2/(2\sigma^2)}$ es par.
- Los términos desplazados están en $\pm 1$ y como $A_{-1}=A_1$ y la varianza es la misma, se cumple:

$$
A_1 e^{-\frac{(\tau-1)^2}{2\sigma^2}} + A_1 e^{-\frac{(\tau+1)^2}{2\sigma^2}}
\quad \Rightarrow \quad
\text{es una función par}.
$$

Por tanto, $R(\tau)$ **sí es par**, lo cual es consistente con autocorrelación de un proceso real WSS.

---

### **(ii) Verificación clave con Wiener–Khinchin: PSD no negativa**

Usamos que la transformada de Fourier (CTFT) de una gaussiana desplazada es:

$$
\mathcal{F}\left\{e^{-\frac{(\tau-\mu)^2}{2\sigma^2}}\right\}
=
\sqrt{2\pi}\,\sigma\;
e^{-\frac{1}{2}\sigma^2\omega^2}\;e^{-j\omega\mu}.
$$

Aplicando esto a cada término:

- Para $k=0$ ($\mu_0=0$):
$$
S_0(\omega)=\sqrt{2\pi}\,\sigma\;A_0\;e^{-\frac{1}{2}\sigma^2\omega^2}.
$$

- Para $k=1$ ($\mu_1=1$):
$$
S_1(\omega)=\sqrt{2\pi}\,\sigma\;A_1\;e^{-\frac{1}{2}\sigma^2\omega^2}e^{-j\omega}.
$$

- Para $k=-1$ ($\mu_{-1}=-1$):
$$
S_{-1}(\omega)=\sqrt{2\pi}\,\sigma\;A_1\;e^{-\frac{1}{2}\sigma^2\omega^2}e^{+j\omega}.
$$

Sumando:

$$
S_x(\omega)=\sqrt{2\pi}\,\sigma\;e^{-\frac{1}{2}\sigma^2\omega^2}
\Big(A_0 + A_1 e^{-j\omega}+A_1 e^{+j\omega}\Big).
$$

Y usando $e^{-j\omega}+e^{j\omega}=2\cos(\omega)$:

$$
S_x(\omega)=\sqrt{2\pi}\,\sigma\;e^{-\frac{1}{2}\sigma^2\omega^2}
\Big(A_0 + 2A_1\cos(\omega)\Big).
$$

Observación: el factor
$$
\sqrt{2\pi}\,\sigma\;e^{-\frac{1}{2}\sigma^2\omega^2}
$$
es **siempre positivo** (para $\sigma>0$). Por tanto, el signo de $S_x(\omega)$ depende solo de:

$$
A_0 + 2A_1\cos(\omega).
$$

Como $\cos(\omega)\in[-1,1]$, el caso más crítico es $\cos(\omega)=-1$:

$$
A_0 + 2A_1(-1)\ge 0
\quad\Longleftrightarrow\quad
A_0 - 2A_1 \ge 0
\quad\Longleftrightarrow\quad
A_0 \ge 2A_1.
$$

---

### **Conclusión del Punto 1**
- $R(\tau)$ **cumple** la simetría esperada de una autocorrelación real (es par).
- La condición **decisiva** de validez como autocorrelación WSS (vía Wiener–Khinchin) es:

$$
\boxed{\,A_0 \ge 2A_1\,}
$$

- Si $A_0 < 2A_1$, entonces existen frecuencias (por ejemplo alrededor de $\omega=\pi$) donde:

$$
A_0 + 2A_1\cos(\omega) < 0
\quad\Rightarrow\quad
S_x(\omega)<0,
$$

lo cual **no es posible** para una PSD, y por tanto $R(\tau)$ **NO** puede ser autocorrelación válida en ese caso.

En los siguientes puntos se verificará esto **numéricamente** en PyTorch (definición de $R(\tau)$, gráficas, FFT para estimar $S_x(\omega)$ y detección de regiones negativas).

## 3) Parte Teórica

### (a) Propiedades básicas

#### **Pregunta**
Verificar si la función:

- Es real.  
- Cumple simetría hermítica (para proceso real: paridad) $$R(\tau)=R(-\tau).$$  
- Presenta máximo en $$\tau=0.$$  

Justifique formalmente.

---

#### **Desarrollo**

Partimos de:

$$
R(\tau)=\sum_{k\in\{-1,0,1\}} A_k \exp\!\left(-\frac{(\tau-\mu_k)^2}{2\sigma_k^2}\right),
\qquad
\mu_k=k,
\qquad
A_k>0,\; A_{-1}=A_1.
$$

Para $$k=\pm1$$, como $$|1|!=1$$, se cumple:

$$
\sigma_{\pm1}^2=\frac{\sigma^2}{|1|!}=\sigma^2,
\qquad
\sigma_0^2=\sigma^2.
$$

Entonces, de forma explícita:

$$
R(\tau)=A_0 \exp\!\left(-\frac{\tau^2}{2\sigma^2}\right)
+ A_1 \exp\!\left(-\frac{(\tau-1)^2}{2\sigma^2}\right)
+ A_1 \exp\!\left(-\frac{(\tau+1)^2}{2\sigma^2}\right).
$$

---

### 1) **¿$$R(\tau)$$ es real?**
Sí. Cada término es de la forma $$A_k \exp(\text{real})$$, porque:

- $$A_k>0$$ es real,
- el exponente $$-\dfrac{(\tau-\mu_k)^2}{2\sigma^2}$$ es real para $$\tau\in\mathbb{R}$$ y $$\sigma>0$$,
- por tanto $$\exp(\text{real})\in\mathbb{R}^+$$.

Luego, suma de reales $$\Rightarrow$$

$$
R(\tau)\in\mathbb{R}\quad \forall \tau.
$$

---

### 2) **Simetría hermítica (paridad para proceso real)**
Para un proceso real WSS debe cumplirse:

$$
R(\tau)=R(-\tau).
$$

Calculamos $$R(-\tau)$$:

$$
R(-\tau)=A_0 \exp\!\left(-\frac{(-\tau)^2}{2\sigma^2}\right)
+ A_1 \exp\!\left(-\frac{(-\tau-1)^2}{2\sigma^2}\right)
+ A_1 \exp\!\left(-\frac{(-\tau+1)^2}{2\sigma^2}\right).
$$

Simplificando:

$$
(-\tau)^2=\tau^2,\qquad
(-\tau-1)^2=(\tau+1)^2,\qquad
(-\tau+1)^2=(\tau-1)^2.
$$

Entonces:

$$
R(-\tau)=A_0 \exp\!\left(-\frac{\tau^2}{2\sigma^2}\right)
+ A_1 \exp\!\left(-\frac{(\tau+1)^2}{2\sigma^2}\right)
+ A_1 \exp\!\left(-\frac{(\tau-1)^2}{2\sigma^2}\right)
=R(\tau).
$$

Por tanto:

$$
\boxed{R(\tau)\ \text{es par} \;\Rightarrow\; R(\tau)=R(-\tau).}
$$

---

### 3) **¿Presenta máximo en $$\tau=0$$? (justificación formal)**

Para una ACF de un proceso WSS real se espera:

$$
R(0)=\mathbb{E}\{x^2(t)\}\ge 0,
\qquad
|R(\tau)|\le R(0).
$$

Aquí verificamos formalmente el comportamiento alrededor de $$\tau=0$$.

---

#### 3.1) $$\tau=0$$ es punto crítico (derivada nula)

Sea:

$$
g(\tau)=\exp\!\left(-\frac{(\tau-\mu)^2}{2\sigma^2}\right).
$$

Entonces:

$$
g'(\tau)= -\frac{\tau-\mu}{\sigma^2}\,g(\tau).
$$

Aplicándolo a $$R(\tau)$$:

$$
\begin{aligned}
R'(\tau)=
&-\frac{A_0}{\sigma^2}\,\tau\,\exp\!\left(-\frac{\tau^2}{2\sigma^2}\right)
-\frac{A_1}{\sigma^2}(\tau-1)\exp\!\left(-\frac{(\tau-1)^2}{2\sigma^2}\right) \\
&-\frac{A_1}{\sigma^2}(\tau+1)\exp\!\left(-\frac{(\tau+1)^2}{2\sigma^2}\right).
\end{aligned}
$$

Evaluando en $$\tau=0$$:

- el primer término vale $$0$$,
- los términos de $$\pm1$$ se cancelan por simetría:

$$
R'(0)=
-\frac{A_1}{\sigma^2}(-1)\exp\!\left(-\frac{1}{2\sigma^2}\right)
-\frac{A_1}{\sigma^2}(1)\exp\!\left(-\frac{1}{2\sigma^2}\right)=0.
$$

Así:

$$
\boxed{R'(0)=0 \;\Rightarrow\; \tau=0 \text{ es un punto crítico}.}
$$

---

#### 3.2) Condición para que sea **máximo local** (segunda derivada)

Para $$g(\tau)$$:

$$
g''(\tau)=g(\tau)\,\frac{(\tau-\mu)^2-\sigma^2}{\sigma^4}.
$$

Entonces:

- término central ($$\mu=0$$):

$$
\left[A_0 \exp\!\left(-\frac{\tau^2}{2\sigma^2}\right)\right]''\Big|_{\tau=0}
=
A_0\cdot 1\cdot \frac{0-\sigma^2}{\sigma^4}
=
-\frac{A_0}{\sigma^2}.
$$

- términos $$\mu=\pm1$$ en $$\tau=0$$:

$$
\exp\!\left(-\frac{(0\mp1)^2}{2\sigma^2}\right)=\exp\!\left(-\frac{1}{2\sigma^2}\right),
\qquad
\frac{(0\mp1)^2-\sigma^2}{\sigma^4}=\frac{1-\sigma^2}{\sigma^4}.
$$

Sumando ambos términos:

$$
R''(0)=
-\frac{A_0}{\sigma^2}
+
2A_1\,\exp\!\left(-\frac{1}{2\sigma^2}\right)\frac{1-\sigma^2}{\sigma^4}.
$$

Para que $$\tau=0$$ sea **máximo local**, se requiere:

$$
R''(0)<0.
$$

Es decir:

$$
-\frac{A_0}{\sigma^2}
+
2A_1\,\exp\!\left(-\frac{1}{2\sigma^2}\right)\frac{1-\sigma^2}{\sigma^4}
<0,
$$

equivalente a:

$$
2A_1\,\exp\!\left(-\frac{1}{2\sigma^2}\right)(1-\sigma^2)
<
A_0\,\sigma^2.
$$

**Caso 1: $$\sigma^2\ge 1$$**  
Entonces $$1-\sigma^2\le 0$$ y el segundo término es $$\le 0$$, por lo que:

$$
R''(0)= -\frac{A_0}{\sigma^2} + (\text{algo }\le 0) < 0
\quad\Rightarrow\quad
\boxed{\sigma^2\ge 1 \ \Rightarrow\ 0 \text{ es máximo local}.}
$$

**Caso 2: $$\sigma^2<1$$**  
Entonces $$1-\sigma^2>0$$ y se necesita una relación entre $$A_0$$ y $$A_1$$:

$$
\boxed{
A_0 >
\frac{2A_1\,\exp\!\left(-\frac{1}{2\sigma^2}\right)(1-\sigma^2)}{\sigma^2}
\ \Rightarrow\
0 \text{ es máximo local}.
}
$$

> Nota: que $$\tau=0$$ sea máximo **local** no garantiza que sea máximo **global**. En una suma de gaussianas, si $$A_1$$ es grande, podrían existir picos cerca de $$\tau=\pm1$$. Esto se comprobará en la parte computacional.

---

#### **Conclusión (a)**
- $$\boxed{R(\tau)\ \text{es real}.}$$
- $$\boxed{R(\tau)=R(-\tau)\ \text{(paridad / simetría hermítica para real)}.}$$
- $$\boxed{R'(0)=0\ \Rightarrow\ \tau=0\ \text{es punto crítico}.}$$
- $$\boxed{\tau=0\ \text{es máximo local}}$$ si:
  - $$\sigma^2\ge 1$$, o
  - $$\sigma^2<1$$ y se cumple la desigualdad anterior.

En la parte computacional se verificará si es máximo global (comparación de picos) y, de forma decisiva, se validará la condición de autocorrelación usando la PSD (Wiener–Khinchin).

### (b) Condición de positividad definida

#### **Pregunta**
Recuerde que una función es ACF válida si y solo si es positivamente definida.

- Enuncie formalmente esta propiedad.  
- Explique su relación con matrices de tipo Toeplitz.

---

#### **Desarrollo**

### 1) Enunciado formal: **positividad definida** (condición de ACF válida)

Sea $$R(\tau)$$ una función definida para $$\tau\in\mathbb{R}$$.  
Decimos que $$R(\tau)$$ es **positivamente definida** si para cualquier entero $$N\ge 1$$, para cualquier conjunto de instantes $$t_1,\dots,t_N\in\mathbb{R}$$ y para cualesquiera coeficientes complejos $$c_1,\dots,c_N\in\mathbb{C}$$, se cumple:

$$
\sum_{i=1}^{N}\sum_{j=1}^{N} c_i\,c_j^{*}\,R(t_i-t_j)\ \ge\ 0.
$$

Esta es la forma estándar de la condición.

**Equivalencia con ACF (WSS):**  
Una función $$R(\tau)$$ es una **función de autocorrelación** de algún proceso estacionario en sentido amplio (WSS) **si y solo si**:

1. $$R(0)\ge 0,$$
2. $$R(\tau)=R^{*}(-\tau)$$ (simetría hermítica),
3. $$R(\tau)$$ es **positivamente definida** (condición anterior).

En señales reales, $$R(\tau)=R(-\tau)$$.

---

### 2) Relación con **matrices de tipo Toeplitz**

Tomemos instantes igualmente espaciados:

$$
t_n = nT,\qquad n=0,1,\dots,N-1.
$$

Entonces los retardos entre muestras son:

$$
t_i-t_j=(i-j)T.
$$

Definimos la matriz:

$$
\mathbf{R}=
\begin{bmatrix}
R(0) & R(T) & R(2T) & \cdots & R((N-1)T)\\
R(-T) & R(0) & R(T) & \cdots & R((N-2)T)\\
R(-2T) & R(-T) & R(0) & \cdots & R((N-3)T)\\
\vdots & \vdots & \vdots & \ddots & \vdots\\
R(-(N-1)T) & R(-(N-2)T) & R(-(N-3)T) & \cdots & R(0)
\end{bmatrix}.
$$

Observa que **cada diagonal es constante**, porque la entrada $$ (i,j) $$ depende solo de $$i-j$$:

$$
[\mathbf{R}]_{ij}=R((i-j)T).
$$

Eso es exactamente la estructura **Toeplitz**.

Además, si $$R(\tau)=R^{*}(-\tau)$$, entonces la matriz es **Hermítica**:

$$
\mathbf{R}=\mathbf{R}^{H}.
$$

---

### 3) Conexión directa: positividad definida ↔ Toeplitz semidefinida positiva

La condición de positividad definida puede reescribirse como una forma cuadrática:

Sea $$\mathbf{c}=[c_0,\dots,c_{N-1}]^{T}$$. Entonces:

$$
\sum_{i=0}^{N-1}\sum_{j=0}^{N-1} c_i\,c_j^{*}\,R((i-j)T)
=
\mathbf{c}^{H}\,\mathbf{R}\,\mathbf{c}.
$$

Por tanto, decir que $$R(\tau)$$ es positivamente definida equivale a exigir que, para todo $$N$$:

$$
\boxed{\mathbf{c}^{H}\mathbf{R}\mathbf{c}\ge 0\ \ \forall\,\mathbf{c}}
\quad \Longleftrightarrow \quad
\boxed{\mathbf{R}\succeq 0}
$$

es decir, que **toda** matriz Toeplitz construida a partir de $$R((i-j)T)$$ sea **semidefinida positiva**.

---

### **Conclusión (b)**

- $$R(\tau)$$ es ACF válida **si y solo si** es **positivamente definida** (cumple la desigualdad doble suma para todo conjunto de tiempos y coeficientes).
- En muestreo uniforme, esa condición se traduce en que la matriz de autocorrelación $$\mathbf{R}$$ es **Toeplitz** y debe ser **semidefinida positiva** (todos sus autovalores $$\ge 0$$).

### (c) Uso del Teorema de Wiener–Khinchin

#### **Pregunta**
Para el análisis espectral:

$$
S(f)=\mathcal{F}\{R(\tau)\}.
$$

Utilice el teorema de Wiener–Khinchin para analizar la condición necesaria sobre $$S(f)$$.

Se recomienda:
- Obtener analíticamente la transformada de Fourier de cada término gaussiano.
- Construir la expresión total del espectro.

Analice:
- ¿Es $$S(f)\ge 0$$ para todo $$f$$?
- ¿Depende esto de las amplitudes relativas $$A_k$$?
- ¿Existen configuraciones de parámetros para las cuales la función deja de ser válida?

No se busca únicamente cálculo, sino interpretación estructural.

---

#### **Desarrollo**

### 1) Teorema de Wiener–Khinchin (condición necesaria para ACF válida)

Si $$R(\tau)$$ es una ACF de un proceso WSS, entonces su transformada de Fourier es la **densidad espectral de potencia (PSD)**:

$$
S(f)=\mathcal{F}\{R(\tau)\},
$$

y debe cumplir la condición:

$$
\boxed{S(f)\ge 0 \quad \forall f.}
$$

Esta condición es necesaria y (con simetría hermítica) también sirve como criterio equivalente de validez (Bochner/Wiener–Khinchin).

---

### 2) Reescritura explícita de $$R(\tau)$$ (para $$k\in\{-1,0,1\}$$)

Como se mostró antes, para $$k=\pm1$$ se tiene $$\sigma_{1}^2=\sigma_{-1}^2=\sigma^2$$, y para $$k=0$$ también $$\sigma_0^2=\sigma^2$$. Luego:

$$
R(\tau)=A_0 \exp\!\left(-\frac{\tau^2}{2\sigma^2}\right)
+ A_1 \exp\!\left(-\frac{(\tau-1)^2}{2\sigma^2}\right)
+ A_1 \exp\!\left(-\frac{(\tau+1)^2}{2\sigma^2}\right).
$$

---

### 3) Transformada de Fourier de **cada término gaussiano**

Usaremos la pareja (con variable de frecuencia $$f$$, y convención $$e^{-j2\pi f \tau}$$):

Si
$$
g(\tau)=\exp\!\left(-\frac{(\tau-\mu)^2}{2\sigma^2}\right),
$$
entonces

$$
\mathcal{F}\{g(\tau)\}
=
\sqrt{2\pi}\,\sigma\;
\exp\!\left(-2\pi^2\sigma^2 f^2\right)\;
\exp\!\left(-j2\pi f \mu\right).
$$

Aplicando a cada término:

- Para $$\mu=0$$:
$$
S_0(f)=\sqrt{2\pi}\,\sigma\;A_0\;\exp\!\left(-2\pi^2\sigma^2 f^2\right).
$$

- Para $$\mu=1$$:
$$
S_{+1}(f)=\sqrt{2\pi}\,\sigma\;A_1\;\exp\!\left(-2\pi^2\sigma^2 f^2\right)\exp(-j2\pi f).
$$

- Para $$\mu=-1$$:
$$
S_{-1}(f)=\sqrt{2\pi}\,\sigma\;A_1\;\exp\!\left(-2\pi^2\sigma^2 f^2\right)\exp(+j2\pi f).
$$

---

### 4) Construcción del espectro total $$S(f)$$

Sumando:

$$
\begin{aligned}
S(f)
&=\sqrt{2\pi}\,\sigma\;\exp\!\left(-2\pi^2\sigma^2 f^2\right)
\Big(A_0 + A_1 e^{-j2\pi f}+A_1 e^{+j2\pi f}\Big)\\[4pt]
&=\sqrt{2\pi}\,\sigma\;\exp\!\left(-2\pi^2\sigma^2 f^2\right)
\Big(A_0 + 2A_1\cos(2\pi f)\Big).
\end{aligned}
$$

Por tanto:

$$
\boxed{
S(f)=\sqrt{2\pi}\,\sigma\;e^{-2\pi^2\sigma^2 f^2}\left(A_0+2A_1\cos(2\pi f)\right).
}
$$

---

### 5) Análisis estructural: ¿cuándo $$S(f)\ge 0$$?

Observación clave:

$$
\sqrt{2\pi}\,\sigma\;e^{-2\pi^2\sigma^2 f^2} > 0 \quad \forall f,\ \sigma>0.
$$

Entonces el signo de $$S(f)$$ depende **solo** del factor:

$$
A_0+2A_1\cos(2\pi f).
$$

Como

$$
\cos(2\pi f)\in[-1,1],
$$

el caso más crítico (mínimo) ocurre cuando:

$$
\cos(2\pi f)=-1
\quad\Rightarrow\quad
A_0+2A_1(-1)=A_0-2A_1.
$$

Por tanto, para garantizar $$S(f)\ge 0$$ para todo $$f$$ se requiere:

$$
\boxed{A_0-2A_1\ge 0 \quad \Longleftrightarrow\quad A_0\ge 2A_1.}
$$

---

### 6) Respuestas a las preguntas de interpretación

#### **(i) ¿Es $$S(f)\ge 0$$ para todo $$f$$?**
- **Sí**, siempre que:
$$
\boxed{A_0\ge 2A_1.}
$$
- **No**, si:
$$
A_0<2A_1,
$$
porque existen frecuencias donde $$\cos(2\pi f)\approx -1$$ (por ejemplo $$f=\tfrac{1}{2},\tfrac{3}{2},\dots$$) y entonces:

$$
S(f)\propto A_0-2A_1<0.
$$

---

#### **(ii) ¿Depende esto de las amplitudes relativas $$A_k$$?**
Sí, totalmente. La “forma” del espectro es un **gaussiano positivo** (envolvente) multiplicado por un término oscilatorio:

$$
A_0+2A_1\cos(2\pi f).
$$

- $$A_0$$ actúa como “nivel base” (offset).
- $$A_1$$ actúa como “profundidad de modulación” (la oscilación por el coseno).

Si la modulación es demasiado grande frente al offset (es decir, $$2A_1>A_0$$), el espectro cruza por debajo de cero y deja de ser PSD.

---

#### **(iii) ¿Existen configuraciones donde la función deja de ser válida?**
Sí. Precisamente cuando:

$$
\boxed{A_0<2A_1.}
$$

En ese caso $$S(f)$$ toma valores negativos en bandas de frecuencia alrededor de donde $$\cos(2\pi f)=-1$$, y por Wiener–Khinchin esto **no puede** corresponder a una densidad espectral de potencia.

---

### **Conclusión (c)**

Por el Teorema de Wiener–Khinchin, una condición necesaria (y aquí decisiva) para que $$R(\tau)$$ sea ACF válida es:

$$
\boxed{S(f)\ge 0\ \forall f.}
$$

Dado que:

$$
S(f)=\sqrt{2\pi}\,\sigma\;e^{-2\pi^2\sigma^2 f^2}\left(A_0+2A_1\cos(2\pi f)\right),
$$

la condición de no negatividad se reduce a:

$$
\boxed{A_0\ge 2A_1.}
$$

Si $$A_0<2A_1$$, el espectro tiene regiones negativas, lo cual rompe la interpretación como PSD y hace que $$R(\tau)$$ **no** sea una ACF válida.

## 4) Parte Computacional (Python)

### **Pregunta**
Se debe:

1. Definir $$R(\tau)$$ en un intervalo suficientemente amplio.  
2. Graficar la función para distintos valores de:
   - $$A_0$$
   - $$A_1$$
   - $$\sigma$$  
3. Calcular numéricamente su transformada de Fourier usando FFT.  
4. Graficar la densidad espectral de potencia estimada.  
5. Analizar si aparecen regiones negativas en el espectro.

Se deben presentar al menos **tres configuraciones distintas** de parámetros y discutir los resultados.

---

### **Desarrollo / Plan de simulación**
- Se define una malla uniforme de $$\tau\in[-T_{\max},T_{\max}]$$ con suficiente resolución.
- Se implementa:

$$
R(\tau)=A_0 e^{-\frac{\tau^2}{2\sigma^2}}
+ A_1 e^{-\frac{(\tau-1)^2}{2\sigma^2}}
+ A_1 e^{-\frac{(\tau+1)^2}{2\sigma^2}}.
$$

- Se estima:

$$
\hat S(f)\approx \text{FFT}\{R(\tau)\},
$$

incluyendo el factor $$\Delta\tau$$ para aproximar la integral.
- Se revisa el valor mínimo:

$$
\min_f \hat S(f)
$$

para decidir si hay regiones negativas.

---

### **Configuraciones evaluadas**
1. **Caso 1 (válido teórico):** $$A_0=2A_1$$  
2. **Caso 2 (inválido teórico):** $$A_0<2A_1$$  
3. **Caso 3 (válido teórico):** $$A_0\gg 2A_1$$ y $$\sigma$$ pequeño (para ver el efecto en el ancho del espectro)

---

### **Discusión de resultados**
- En los **casos válidos** (1 y 3), el espectro estimado puede mostrar valores mínimos ligeramente negativos del orden de $$10^{-7}$$.  
  Esto se interpreta como **error numérico** (discretización/FFT), no como violación estructural real.

- En el **caso inválido** (2), la negatividad es del orden de $$10^{-2}$$, es decir **mucho mayor** que el error numérico típico.  
  Esto confirma la interpretación estructural del Teorema de Wiener–Khinchin: si $$A_0<2A_1$$, el término oscilatorio domina y el espectro cruza a valores negativos, por lo que $$R(\tau)$$ **no puede ser ACF válida**.

- Efecto de $$\sigma$$:
  - Al disminuir $$\sigma$$, $$R(\tau)$$ se hace más “angosta” en tiempo.
  - Por dualidad, $$\hat S(f)$$ se hace más “ancha” en frecuencia (más contenido espectral).
  - Pero la condición de no negatividad sigue dependiendo del balance $$A_0$$ vs $$A_1$$ (no del ancho).

```python
import numpy as np
import torch
import matplotlib.pyplot as plt

# -----------------------------
# 4) Parte Computacional (Python - PyTorch)
# -----------------------------

def R_tau(tau: torch.Tensor, A0: float, A1: float, sigma: float) -> torch.Tensor:
    """Autocorrelación como suma de gaussianas para k in {-1,0,1}."""
    return (A0 * torch.exp(-(tau**2) / (2*sigma**2)) +
            A1 * torch.exp(-((tau - 1.0)**2) / (2*sigma**2)) +
            A1 * torch.exp(-((tau + 1.0)**2) / (2*sigma**2)))

def estimate_S_f_fft(tau: torch.Tensor, R: torch.Tensor):
    """
    Estima S(f) ≈ FFT{R(tau)} con muestreo uniforme en tau.
    Devuelve: f (Hz) y S_est (real) centrados (fftshift).
    """
    d_tau = float(tau[1] - tau[0])
    N = tau.numel()

    S = torch.fft.fftshift(torch.fft.fft(torch.fft.ifftshift(R))) * d_tau
    f = torch.fft.fftshift(torch.fft.fftfreq(N, d=d_tau))
    return f, torch.real(S)

# 1) Definir tau en intervalo amplio
Tmax = 10.0
N = 2**14
tau = torch.linspace(-Tmax, Tmax, N)

# Tres configuraciones
configs = [
    {"name": "Caso 1 (válido): A0=2A1", "A0": 2.0, "A1": 1.0, "sigma": 1.0},
    {"name": "Caso 2 (inválido): A0<2A1", "A0": 1.0, "A1": 1.0, "sigma": 1.0},
    {"name": "Caso 3 (válido): A0>>2A1, sigma pequeño", "A0": 4.0, "A1": 1.0, "sigma": 0.5},
]

# 2) Graficar R(tau)
plt.figure()
for cfg in configs:
    R = R_tau(tau, cfg["A0"], cfg["A1"], cfg["sigma"])
    plt.plot(tau.numpy(), R.numpy(), label=cfg["name"])
plt.xlabel(r'$\tau$')
plt.ylabel(r'$R(\tau)$')
plt.title(r'Autocorrelación $R(\tau)$ para distintas configuraciones')
plt.legend()
plt.grid(True)
plt.show()

# 3-4) FFT para estimar S(f) y graficar PSD
plt.figure()
neg_report = []
for cfg in configs:
    R = R_tau(tau, cfg["A0"], cfg["A1"], cfg["sigma"])
    f, S_est = estimate_S_f_fft(tau, R)

    S_min = float(S_est.min())
    neg_report.append((cfg["name"], S_min))

    f_np = f.numpy()
    S_np = S_est.numpy()
    mask = np.abs(f_np) <= 3.0
    plt.plot(f_np[mask], S_np[mask], label=f'{cfg["name"]} (min={S_min:.3e})')

plt.xlabel(r'$f$ (Hz)')
plt.ylabel(r'$\hat{S}(f)$')
plt.title(r'PSD estimada $\hat{S}(f)$ mediante FFT')
plt.legend()
plt.grid(True)
plt.show()

# 5) Reporte de negatividad
print("=== Reporte de valores mínimos de la PSD estimada ===")
for name, S_min in neg_report:
    print(f"{name}: min(Ŝ(f)) = {S_min:.6e}")
```

## 5) Preguntas de análisis

### **Pregunta 1.** ¿Toda suma de funciones gaussianas genera una ACF válida?

**Respuesta:** No necesariamente.  
Que $$R(\tau)$$ sea una suma de gaussianas **no garantiza** que sea una ACF válida. Para ser ACF de un proceso WSS debe cumplir, además de simetría hermítica, la condición estructural clave:

$$
S(f)=\mathcal{F}\{R(\tau)\}\ge 0 \quad \forall f.
$$

En nuestro caso:

$$
S(f)=\sqrt{2\pi}\,\sigma\,e^{-2\pi^2\sigma^2 f^2}\left(A_0+2A_1\cos(2\pi f)\right).
$$

Como el factor gaussiano es siempre positivo, la validez depende de que:

$$
A_0+2A_1\cos(2\pi f)\ge 0\quad \forall f,
$$

lo cual exige el umbral:

$$
\boxed{A_0\ge 2A_1.}
$$

Por tanto, incluso siendo suma de gaussianas, si $$A_0<2A_1$$ aparece espectro negativo y la función deja de ser ACF válida.

---

### **Pregunta 2.** ¿Qué papel juegan las gaussianas desplazadas respecto al término central?

Las gaussianas desplazadas (en $$\tau=\pm 1$$) introducen **estructura oscilatoria en frecuencia**.

- En el dominio de $$\tau$$, agregan “correlación” a retardos cercanos a $$\pm1$$ (picos laterales).
- En el dominio espectral, cada desplazamiento produce un factor complejo:

$$
e^{-j2\pi f\mu},
$$

y al sumar simétricamente $$\mu=\pm1$$ se obtiene el término real:

$$
e^{-j2\pi f}+e^{+j2\pi f}=2\cos(2\pi f).
$$

Por eso, las gaussianas desplazadas **modulan** el espectro del término central:

$$
S(f)\propto A_0 + 2A_1\cos(2\pi f).
$$

Interpretación: el término central $$A_0$$ “sostiene” el espectro con un piso positivo, mientras que los términos desplazados generan “ondulaciones” (modulación) que pueden llevarlo a cruzar por debajo de cero si son demasiado fuertes.

---

### **Pregunta 3.** ¿Existe un umbral en las amplitudes relativas que cambie la validez?

Sí, y es precisamente el criterio estructural más importante del taller.

Como $$\cos(2\pi f)\in[-1,1]$$, el mínimo de:

$$
A_0+2A_1\cos(2\pi f)
$$

ocurre cuando $$\cos(2\pi f)=-1$$. Entonces:

$$
\min_f \left(A_0+2A_1\cos(2\pi f)\right)=A_0-2A_1.
$$

La condición para que $$S(f)\ge 0$$ para todo $$f$$ es:

$$
\boxed{A_0-2A_1\ge 0 \quad\Longleftrightarrow\quad A_0\ge 2A_1.}
$$

- Si $$A_0\ge 2A_1$$ → espectro no negativo → **ACF válida**.
- Si $$A_0<2A_1$$ → espectro negativo en ciertas bandas → **ACF inválida**.

---

### **Pregunta 4.** ¿Qué interpretación física tendría el proceso si la función fuera válida?

Si $$R(\tau)$$ fuera válida, se interpretaría como la autocorrelación de un proceso WSS cuyo contenido espectral está dado por:

$$
S(f)=\sqrt{2\pi}\,\sigma\,e^{-2\pi^2\sigma^2 f^2}\left(A_0+2A_1\cos(2\pi f)\right).
$$

Interpretación física (intuitiva):

- El factor
$$
e^{-2\pi^2\sigma^2 f^2}
$$
es una envolvente gaussiana: sugiere un proceso con energía/ potencia concentrada alrededor de bajas frecuencias (tipo “banda base” suave), con ancho controlado por $$\sigma$$.

- El factor
$$
A_0+2A_1\cos(2\pi f)
$$
introduce una modulación periódica del espectro: indica que el proceso tendría una “huella” de dependencia temporal alrededor de retardos $$\pm 1$$.  
En términos simples: además de correlación fuerte en $$\tau=0$$, habría una componente de “memoria” aproximadamente a 1 unidad de tiempo.

Una lectura práctica: podría modelar un proceso con “eco” o persistencia a un retardo fijo (simétrico), pero suficientemente débil para no romper la no negatividad del espectro.

---

### **Pregunta 5.** ¿Qué sucede cuando el espectro toma valores negativos?

Si $$S(f)$$ toma valores negativos, ocurre un problema estructural fundamental:

- $$S(f)$$ deja de poder interpretarse como **densidad espectral de potencia (PSD)**, porque la potencia por frecuencia no puede ser negativa.
- Por el Teorema de Wiener–Khinchin (y Bochner), esto implica:

$$
\boxed{R(\tau)\ \text{NO es positivamente definida} \ \Rightarrow\ R(\tau)\ \text{NO puede ser ACF válida}.}
$$

En la práctica computacional, una pequeña negatividad del orden de $$10^{-7}$$ puede deberse a error numérico (FFT, truncamiento, discretización).  
Pero negatividades claras (por ejemplo $$10^{-2}$$ como en el caso $$A_0<2A_1$$) indican una violación real del criterio:

$$
S(f)\ngeq 0.
$$

**Consecuencia:** no existe ningún proceso WSS real que tenga esa $$R(\tau)$$ como autocorrelación.
